```python
import tensorflow
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile

pd.set_option('display.max_rows', None)
```

```python
dataset = pd.read_csv("dataset/bird_songs_metadata.csv")[["id", "name", "filename"]]

dataset.head()
```

```python
print(set(dataset["name"]))

print(dataset)
```

```python
# Get a spectrogram
sample_rate, samples = wavfile.read('dataset/wavfiles/557838-9.wav')
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)

plt.pcolormesh(times, frequencies, np.log(spectrogram))
#plt.imshow(spectrogram)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()


```

```python
# Get a spectrogram
sample_rate, samples = wavfile.read('dataset/wavfiles/11713-10.wav')
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)


BIRD_NAME = "BIRD NAME"
BIRD_ID = "BIRD ID"
BIRD_FILENAME = "BIRD FILENAME"

plt.pcolormesh(times, frequencies, np.log(spectrogram))
#plt.imshow(spectrogram)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')



stats = f"bird name: {BIRD_NAME}\n" \
		f"bird id: {BIRD_ID}\n" \
		f"bird filename: {BIRD_FILENAME}"

props = dict(
	boxstyle="round", facecolor="grey", alpha=0.15
)

plt.text(
	1.03,
	0.98,
	stats,
	transform=plt.gca().transAxes,
	fontsize=7,
	verticalalignment="top",
	bbox=props,
)

plt.show()
```
